GitHub Copilot is an AI pair programmer that helps you write code faster and with less effort. It works by:

- Suggesting whole lines or blocks of code as you type
- Converting comments into working code
- Helping you understand and work with unfamiliar libraries
- Offering multiple suggestions for you to choose from
- Learning from your coding style and preferences

````python
# Create a scatter plot using matplotlib with the following features:
# - Random data points
# - Custom colors based on point values
# - A trend line
# - Proper labels and title
# - A legend
# - Proper

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Generate random data points
np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)

# Assign custom colors based on y values
colors = y - y.min()

# Create the scatter plot
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, c=colors, cmap='viridis', label='Data Points')

# Add a trend line with an initial slope
initial_slope = 2
m, b = np.polyfit(x, y, 1)
trend_line, = ax.plot(x, initial_slope * x + b, color='red', label='Trend Line')

# Add labels, title, and legend
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Scatter Plot with Adjustable Trend Line')
ax.legend()

# Add a slider for adjusting the slope
ax_slider = plt.axes([0.2, 0.01, 0.65, 0.03], facecolor='lightgoldenrodyellow')
slope_slider = Slider(ax_slider, 'Slope', 0.1, 5.0, valinit=initial_slope)

# Update function for the slider
def update(val):
    slope = slope_slider.val
    trend_line.set_ydata(slope * x + b)
    fig.canvas.draw_idle()

slope_slider.on_changed(update)

# Show the plot
plt.colorbar(scatter, label='Color Scale')
plt.show()
````

```python
# Create a FastAPI application that:
# 1. Implements a RESTful API for managing patient data
# 2. Provides endpoints for adding, retrieving, and updating patient records
# 3. Includes data validation and error handling
# 4. Implements authentication for secure access
# 5. Returns data in JSON format with proper status codes
```


````python
from fastapi import FastAPI, HTTPException, Depends
from pydantic import BaseModel
from typing import List
from fastapi.security import HTTPBasic, HTTPBasicCredentials

app = FastAPI()
security = HTTPBasic()

# In-memory database for patient records
patients = {}

# Pydantic model for patient data
class Patient(BaseModel):
    id: int
    name: str
    age: int
    condition: str

# Authentication dependency
def authenticate(credentials: HTTPBasicCredentials = Depends(security)):
    if credentials.username != "admin" or credentials.password != "password":
        raise HTTPException(status_code=401, detail="Invalid credentials")

# Endpoint to add a new patient
@app.post("/patients/", status_code=201, dependencies=[Depends(authenticate)])
def add_patient(patient: Patient):
    if patient.id in patients:
        raise HTTPException(status_code=400, detail="Patient ID already exists")
    patients[patient.id] = patient
    return {"message": "Patient added successfully"}

# Endpoint to retrieve a patient by ID
@app.get("/patients/{patient_id}", response_model=Patient, dependencies=[Depends(authenticate)])
def get_patient(patient_id: int):
    if patient_id not in patients:
        raise HTTPException(status_code=404, detail="Patient not found")
    return patients[patient_id]

# Endpoint to update a patient's record
@app.put("/patients/{patient_id}", dependencies=[Depends(authenticate)])
def update_patient(patient_id: int, updated_patient: Patient):
    if patient_id not in patients:
        raise HTTPException(status_code=404, detail="Patient not found")
    patients[patient_id] = updated_patient
    return {"message": "Patient updated successfully"}

# Run the application using: uvicorn filename:app --reload
````

### ✏️ Exercise Create a function that processes medical image data:
   - Accept a numpy array representing an image
   - Apply basic image processing (normalization, filtering)
   - Return the processed image
   - Include proper error handling

````python
# Import necessary libraries
import numpy as np
from scipy.ndimage import gaussian_filter

def process_medical_image(image):
    """
    Processes a medical image by normalizing and applying a Gaussian filter.

    Parameters:
        image (numpy.ndarray): Input image as a numpy array.

    Returns:
        numpy.ndarray: Processed image.

    Raises:
        ValueError: If the input is not a numpy array or is empty.
    """
    if not isinstance(image, np.ndarray):
        raise ValueError("Input must be a numpy array.")
    if image.size == 0:
        raise ValueError("Input image cannot be empty.")
    
    # Normalize the image to range [0, 1]
    normalized_image = (image - np.min(image)) / (np.max(image) - np.min(image))
    
    # Apply a Gaussian filter for smoothing
    processed_image = gaussian_filter(normalized_image, sigma=1)
    
    return processed_image

# Example usage
if __name__ == "__main__":
    # Create a sample image (e.g., a 5x5 array)
    sample_image = np.random.rand(5, 5) * 255  # Simulating a grayscale image
    try:
        processed_image = process_medical_image(sample_image)
        print("Processed Image:")
        print(processed_image)
    except ValueError as e:
        print(f"Error: {e}")
````


### ✏️ Exercise: Write a class for managing patient records
   - Include methods for adding, updating, and retrieving records
   - Implement data validation
   - Add methods for basic statistical analysis

````python
class PatientRecords:
    """
    A class to manage patient records with methods for adding, updating, 
    retrieving records, and performing basic statistical analysis.
    """
    def __init__(self):
        self.records = {}

    def add_record(self, patient_id, name, age, weight, height):
        """
        Adds a new patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            name (str): Name of the patient.
            age (int): Age of the patient.
            weight (float): Weight of the patient in kg.
            height (float): Height of the patient in meters.

        Raises:
            ValueError: If any input is invalid.
        """
        if not patient_id or not isinstance(patient_id, str):
            raise ValueError("Patient ID must be a non-empty string.")
        if not name or not isinstance(name, str):
            raise ValueError("Name must be a non-empty string.")
        if age <= 0 or not isinstance(age, int):
            raise ValueError("Age must be a positive integer.")
        if weight <= 0 or not isinstance(weight, (int, float)):
            raise ValueError("Weight must be a positive number.")
        if height <= 0 or not isinstance(height, (int, float)):
            raise ValueError("Height must be a positive number.")
        
        self.records[patient_id] = {
            "name": name,
            "age": age,
            "weight": weight,
            "height": height
        }

    def update_record(self, patient_id, **kwargs):
        """
        Updates an existing patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.
            kwargs: Fields to update (e.g., name, age, weight, height).

        Raises:
            KeyError: If the patient ID does not exist.
            ValueError: If any input is invalid.
        """
        if patient_id not in self.records:
            raise KeyError(f"No record found for patient ID: {patient_id}")
        
        for key, value in kwargs.items():
            if key in self.records[patient_id]:
                if key in ["age", "weight", "height"] and value <= 0:
                    raise ValueError(f"{key.capitalize()} must be a positive number.")
                self.records[patient_id][key] = value

    def get_record(self, patient_id):
        """
        Retrieves a patient record.

        Parameters:
            patient_id (str): Unique identifier for the patient.

        Returns:
            dict: Patient record.

        Raises:
            KeyError: If the patient ID does not exist.
        """
        if patient_id not in self.records:
            raise KeyError(f"No record found for patient ID: {patient_id}")
        return self.records[patient_id]

    def calculate_average_age(self):
        """
        Calculates the average age of all patients.

        Returns:
            float: Average age.
        """
        if not self.records:
            return 0
        total_age = sum(record["age"] for record in self.records.values())
        return total_age / len(self.records)

    def calculate_average_bmi(self):
        """
        Calculates the average BMI of all patients.

        Returns:
            float: Average BMI.
        """
        if not self.records:
            return 0
        total_bmi = 0
        for record in self.records.values():
            weight = record["weight"]
            height = record["height"]
            bmi = weight / (height ** 2)
            total_bmi += bmi
        return total_bmi / len(self.records)

# Example usage
if __name__ == "__main__":
    records = PatientRecords()
    records.add_record("P001", "John Doe", 30, 70, 1.75)
    records.add_record("P002", "Jane Smith", 25, 60, 1.65)
    print("Patient P001:", records.get_record("P001"))
    records.update_record("P001", weight=72)
    print("Updated Patient P001:", records.get_record("P001"))
    print("Average Age:", records.calculate_average_age())
    print("Average BMI:", records.calculate_average_bmi())
````

### ✏️ Exercise: Create a data visualization dashboard
   - Use pandas for data manipulation
   - Create multiple types of plots using matplotlib
   - Add interactive elements

````python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Sample data
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 40, 45],
    "Height": [1.65, 1.80, 1.75, 1.70, 1.60],
    "Weight": [55, 75, 68, 72, 50],
    "City": ["New York", "Los Angeles", "New York", "Chicago", "Los Angeles"]
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Function to display a bar chart of a selected column
def plot_bar_chart(column):
    """
    Plots a bar chart for the selected column.

    Parameters:
        column (str): Column name to plot.
    """
    plt.figure(figsize=(8, 5))
    df[column].value_counts().plot(kind="bar", color="skyblue")
    plt.title(f"Bar Chart of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()

# Function to display a scatter plot of two selected columns
def plot_scatter(x_column, y_column):
    """
    Plots a scatter plot for the selected columns.

    Parameters:
        x_column (str): Column name for the x-axis.
        y_column (str): Column name for the y-axis.
    """
    plt.figure(figsize=(8, 5))
    plt.scatter(df[x_column], df[y_column], color="orange")
    plt.title(f"Scatter Plot of {x_column} vs {y_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()

# Interactive bar chart
print("Interactive Bar Chart:")
interact(plot_bar_chart, column=widgets.Dropdown(options=df.columns, description="Column:"))

# Interactive scatter plot
print("\nInteractive Scatter Plot:")
interact(plot_scatter, 
         x_column=widgets.Dropdown(options=df.columns, description="X Column:"),
         y_column=widgets.Dropdown(options=df.columns, description="Y Column:"))
````
